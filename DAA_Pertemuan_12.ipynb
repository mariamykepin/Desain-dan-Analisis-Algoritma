```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
vertices1 = range (1,10)
edges1 = [(7, 2), (2, 3), (7, 4), (4, 5), (7, 3), (7, 5), (1, 6), (1, 7), (2, 8), (2, 9)]
G = nx.Graph()
G.add_nodes_from (vertices1)
G.add_edges_from (edges1)
nx.draw (G, with_labels = True, node_color = 'yellow', node_size = 800)
```

```python
  nx.degree_centrality (G)
```

```python
nx.betweenness_centrality (G)
```

```python
nx.closeness_centrality (G)
```

```python
centrality = nx.eigenvector_centrality (G)
sorted ((v, '{:0.2f}'.format(c)) for v, c in centrality.items ())
```

```python
G_fb = nx.read_edgelist ("facebook.combined.txt", create_using = nx.Graph(), nodetype = int)
```

```python
print (nx.info (G_fb))
```

```python
pos = nx.spring_layout (G_fb)
betCent = nx.betweenness_centrality (G_fb, normalized = True, endpoints = True)
node_color = [20000.0 * G_fb.degree (v) for v in G_fb]
node_size = [v * 10000 for v in betCent.values ()]

plt.figure (figsize = (20, 20))
nx.draw_networkx (G_fb, pos = pos, with_labels = False,
                  node_color = node_color,
                  node_size = node_size
                  )

plt.axis ('off')
```
